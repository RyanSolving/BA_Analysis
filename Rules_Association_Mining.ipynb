```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
tfidf = pd.read_csv('/content/raw_tfidf.csv')
```

```python
np.median(tfidf.values)
```

```python
tfidf = tfidf.fillna(0)
```

# Rules mining

```python
threshold = tfidf.replace(0,np.nan).median(axis=1)
```

```python
concated_keywords = pd.DataFrame(index=tfidf.index)
```

```python
for idx in tfidf.index:
  row = tfidf.loc[idx]
  row_threshold = threshold[idx]
  keywords = row[row>row_threshold].index.tolist()
  concated_keywords.loc[idx,'Filtered Keywords'] = ' '.join(keywords)
```

```python
te = TransactionEncoder()
te_array = te.fit_transform(concated_keywords['Filtered Keywords'].str.split(' '))
df_encoded = pd.DataFrame(te_array,
                          columns=te.columns_,
                          index=concated_keywords.index)
```

```python
df_encoded_dropbiz = df_encoded.drop(columns=['business','ability'],axis=1)
```

```python
frequent_items = apriori(df_encoded_dropbiz, min_support=0.1, use_colnames=True)
```

```python
rules = association_rules(frequent_items, metric="confidence", min_threshold=0)
```
